```python
import pandas as pd #판다스 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read_csv() 함수를 사용해 csv파일 읽기
df_clean = pd.read_csv('/content/drive/MyDrive/와인 리뷰/wine_review_clean.csv', nrows =317221)
df_clean
```

```python
wine_name = df_clean['wine'].tolist()
```

```python
wine_name
```

```python
wine_name_revise = []

for name in wine_name:
  lst = name.split(' ')
  name_revise = ''
  tmp_num = 0
  tmp_brak = len(lst)
  num = 0
  for word in lst:
    if word == '':
        num += 1
        continue
    if word[0] == '1' or word[0] == '2':
      tmp_num = num +1
    if word[0] == '(':
      tmp_brak = num
    num += 1
  for i in range(tmp_num, tmp_brak):
    name_revise = name_revise + lst[i] +' '
  name_revise = name_revise.strip()
  wine_name_revise.append(name_revise)
```

```python
wine_name_revise
```

```python
df_clean['wine'] = wine_name_revise
```
